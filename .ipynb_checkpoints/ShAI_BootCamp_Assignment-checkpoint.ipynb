### 3-Data Cleaning
* Handle missing data 

```python
# first we see the number of missing values of every column
df.isnull().sum()
```

#### For filling Base Pay column:
#### Total Pay = Base Pay + Overtime Pay + Other Pay
#### So Base Pay = Total Pay - (Overtime Pay + Other Pay)

```python
# Filling according to the equation
df['BasePay'].fillna(df['TotalPay'] - (df['OvertimePay'] + df['OtherPay']), inplace=True)
```

#### For filling Benefits column:
#### Total Pay Benefits = Benefits + Base Pay + Overtime Pay + Other Pay
#### So Benefits = Total Pay Benefits - (Base Pay + Overtime Pay + Other Pay)

```python
# Filling according to the equation
df['Benefits'].fillna(df['TotalPayBenefits'] - (df['BasePay'] + df['OvertimePay'] + df['OtherPay']), inplace=True)
```

```python
print("Missing values in each column after filling:")
print(df.isnull().sum())
```

#### we can see that the Columns (Notes, Status) are all empty So let's remove them along with the ID column as we don't need it


```python
df=df.drop(['Notes','Status','Id'],axis=1)
```

```python
# check for duplicated names
df['EmployeeName'].duplicated().sum()
```

### 4-Basic Data Visualization

* Create histograms or bar charts to visualize the distribution of salaries
* Use pie charts to represent the proportion of employees in different departments

```python
# Divide the salaries into 10 bins
bins = np.linspace(min(df['TotalPay']), max(df['TotalPay']), 20)

# Create the histogram
plt.figure(figsize = (12, 6))
plt.hist(df['TotalPay'], bins=bins, edgecolor='black')

# Set the title and labels
plt.title('Distribution of Employee Salaries')
plt.xlabel('Salary Range')
plt.ylabel('Number of Employees')

# Show the plot
plt.show()
```

```python
df['JobTitle'].value_counts()[:10].index
```

```python
# len(df['JobTitle'].value_counts())
# there is 2159 unique job title in the company so I'm making the pie for only top 10

# Create the pie chart
plt.figure(figsize = (12, 6))
plt.pie(df['JobTitle'].value_counts()[:10],labels=df['JobTitle'].value_counts()[:10].index, autopct='%1.1f%%')

# Set the title
plt.title('Proportion of Employees in Different Departments')
```

### 5-Grouped Analysis
* Group the data by one or more columns and calculate summary statistics for each group
* Compare the average salaries across different groups.

```python
grouped_salary_jobtitle = df.groupby('JobTitle').mean()['TotalPay'].sort_values(ascending=False)

# Create the histogram
plt.figure(figsize = (12, 6))
plt.bar(grouped_salary_jobtitle.index[:20], grouped_salary_jobtitle.values[:20])

# Set the title and labels
plt.title('Mean Salaries by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Mean Salary')

# Rotate the x-axis labels by 90 degrees for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

#### The relationship between the 'BasePay' of the employer and his 'Benefits' ?

### 6- Correlation Analysis

```python
# correlation heatmap for numeric features in the dataset
selected_columns = df.select_dtypes(include=['number'])
correlation_matrix = selected_columns.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Factors vs. Monthly Balance')
plt.show()
```

* We can see that there is a good correlation between the 'TotalPay' and the ['BasePay', 'OvertimePay', 'Benefits']
* There is a low correlation between the 'Benefits' & 'OtherPay' and 'Benefits' & 'OvertimePay' but a good one with 'BasePay
